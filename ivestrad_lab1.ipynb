### Problem 4: Countries Dataset

1. Load the data from the `countries.csv` file into a Pandas data frame.

````python
countries = pd.read_csv('countries.csv')
countries
````

2. What is the overall life expectancy across in the world?

  Hint: You'll need to take a weighted average of life expectancy using populations as weights.

````python
countries['product'] = countries['life_expectancy'] * countries['population']
product = pd.DataFrame(countries['product'])
pd.merge(countries, product)
countries.head()
````

````python
countries['product'].sum() / countries['population'].sum()
````

3. Create a dataframe containing 10 countries with the highest population.

````python
countries_highpop = countries.sort_values('population')
countries_highpop[:10]
````

4. Add a new column in `countries_df` to record the overall GDP per country (product of population & per capita GDP).


````python
countries['GDP'] = countries['population'] * countries['gdp_per_capita']
gdp_df = pd.DataFrame(countries['GDP'])
countries = pd.merge(countries, gdp_df)
countries
````

5. Create a dataframe containing 10 countries with the lowest GDP per capita, among the countries with population greater than 100 million.

````python
pop100 = countries['population'] > 100000000
countries_pop100 = countries[pop100]
countries_pop100 = countries_pop100.sort_values('gdp_per_capita')
countries_pop100 = countries_pop100[:10]
countries_pop100
````
